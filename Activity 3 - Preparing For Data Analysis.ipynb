```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("PeoplesFavourites-1.csv")
```

```python
df.sample(25)
```

#### Anonymize the data by removing the customer names (first and last name) replacing with a unique identifier.

```python
# Anonymyze  First Name and Last Name by creating a unique Id and drop First Name and Last Name
df['Name_Id'] = (df['First Name'] + '_' + df['Last Name']).astype('category').cat.codes
df.drop(columns=['First Name','Last Name'], inplace=True)
```

```python
# anonymyze the email address
df['Email'] = df['Name_Id'].map(str)+'@'+ df['Email'].str.extract("@(.*)")[0]
```

```python
df.sample(25)
```

#### A number of different symbols (including space) have been used for null values, select and apply a consist ant approach to their representation.
    

```python
df = df.replace(['\b','0', 'none'], np.nan)
df = df.replace(0, -1)
```

```python
df.sample(25)
```

#### Remove any customer data that has missing data in any of the following first name, last name and email.
    

```python
index = df[ (df['Email'].isnull()) | (df['Name_Id'] == -1)].index
df.drop(index , inplace=True)
```

```python
df[df['Email'].isnull() | df['Name_Id'] == -1]
```

#### export the data to a new file

```python
df.to_csv('PeoplesFavourites-clean.csv')
```
